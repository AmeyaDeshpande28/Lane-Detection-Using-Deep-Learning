```python
import glob
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt

#Define Size of images
height = 720
width = 1280
imdepth = 3
dataShape = imwidth*imheight

#Number Of classes in labels
classes = 2


dataPath = './LaneDetection/seg_data/train/'
labelPath = './LaneDetection/seg_data/train/Label/'

dataExt = '.jpg'
labelExt = '.jpg'

savePath = './LaneDetection/seg_data/'
dataFileName = 'train_data_new_noRollAxis.npy'
labelFileName = 'train_label_new_noRollAxis.npy'


def sampleMeanStd(img):
    
    img  = img.astype("float32")
    
    b_ch=np.mean(img[:,:,0])
    g_ch=np.mean(img[:,:,1])
    r_ch=np.mean(img[:,:,2])

    #Individual channel-wise mean subtraction
    img -= np.array((b_ch,g_ch,r_ch))
    
    b_ch=np.std(img[:,:,0])
    g_ch=np.std(img[:,:,1])
    r_ch=np.std(img[:,:,2])
    
    img /= np.array((b_ch,g_ch,r_ch))
    
    return img



def prepareDataSet_noRollAxis():

    labelpaths = sorted(glob.glob(labelPath+"*"+labelExt))
    
    #Create Empty Lists to store Image and Label Data
    data = []
    label = []
    imgNames = []
        
    for i in range(len(labelpaths)):
        
        tlp = labelpaths[i]
        tileno = os.path.split(tlp)[-1].split(labelExt)[0]
        tdp = dataPath+tileno+dataExt
                
        im = cv2.imread(tdp)                     
        lab = cv2.imread(tlp)
        
        lab = (lab[:,:,0]>200).astype('uint8')
        
#         cv2.imshow('im', im)
#         cv2.imshow('lab', np.uint8(255*lab))
#         cv2.waitKey(2000)

        data.append((im)) 
        label.append(((lab)))
        imgNames.append(tdp)
        
        print('\n'+ str(i)+'/'+str(len(labelpaths))+' : '+tdp)
    else:
        print("error: "+tdp)

    return data,label,imgNames   
```

```python
data,label,imgNames = prepareDataSet_noRollAxis()

#Store Data to the directory
np.save(savePath+dataFileName, np.array(data))
np.save(savePath+labelFileName, np.array(label))
np.save(savePath+'fileNames.npy', np.array(imgNames))

print('Dataset shape :', np.shape(data))
print('Label shape :', np.shape(label))

print ("Done")
```
